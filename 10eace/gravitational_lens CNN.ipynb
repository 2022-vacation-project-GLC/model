```python
from astropy.table import Table
import pyfits as fits
import numpy as np
import torch

# use GPU
use_cuda = True
device = torch.device("cuda" if use_cuda else "cpu")

# loads the catalog
cat = Table.read('/content/drive/MyDrive/CRA vacation_project/classifications.csv')

ims = np.zeros((20000, 1, 101, 101))

# loads the images
for i, id in enumerate(cat['ID']):
  try:
    ims[i] = fits.getdata('/content/drive/MyDrive/CRA vacation_project/Data.0/Public/Band1/imageEUC_VIS-'+str(id)+'.fits')
  except IOError:
    ims[i] = fits.getdata('/content/drive/MyDrive/CRA vacation_project/Data.1/Public/Band1/imageEUC_VIS-'+str(id)+'.fits')

# Concatenate images to catalog
cat['image'] = ims

cat.write('catalogs.hdf5', path='\space', append=True)
```
